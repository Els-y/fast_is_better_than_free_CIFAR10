```python
import numpy as np
import matplotlib.pyplot as plt
```

CIFAR10 干净样本准确率以及 PGD-50 (epsilon=8/255) 的准确率

```python
fgsm_x = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
fgsm_clean = [0.6336, 0.7411, 0.7810, 0.8049, 0.8241, 0.8325, 0.8417, 0.8452, 0.8478, 0.8518]
fgsm_adv = [0.3715, 0.43, 0.4479, 0.4566, 0.4625, 0.4572, 0.4577, 0.4669, 0.4607, 0.4570]
fgsm_time = [110, 221, 327, 436, 542, 659, 756, 858, 971, 1080]

pgd_x = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
pgd_clean = [0.5910, 0.7026, 0.7475, 0.7790, 0.7964, 0.8072, 0.8149, 0.8209, 0.8255, 0.8273]
pgd_adv = [0.3738, 0.4526, 0.4705, 0.4838, 0.4994, 0.4989, 0.5042, 0.5094, 0.5103, 0.5158]
pgd_time = [404, 806, 1205, 1610, 2025, 2402, 2768, 3197, 3621, 4038]

free_x = list(range(8, 104, 8))
free_clean = [0.4541, 0.5374, 0.5982, 0.6488, 0.6977, 0.7149, 0.7364, 0.7513, 0.7681, 0.7677, 0.7770, 0.7822]
free_adv = [0.2856, 0.3248, 0.3563, 0.3702, 0.3973, 0.4161, 0.4228, 0.4369, 0.4399, 0.4537, 0.4543, 0.4559]
free_time = [101, 200, 296, 395, 489, 588, 679, 780, 874, 979, 1076, 1155]
```

## 准确率

```python
plt.figure(figsize=(8, 5))

plt.plot(fgsm_x, fgsm_clean, color='red', label='fgsm_clean', linestyle='-', marker='s')
plt.plot(fgsm_x, fgsm_adv, color='red', label='fgsm_adv', linestyle=':', marker='s')

plt.plot(pgd_x, pgd_clean, color='green', label='pgd_clean', linestyle='-', marker='^')
plt.plot(pgd_x, pgd_adv, color='green', label='pgd_adv', linestyle=':', marker='^')

plt.plot(free_x, free_clean, color='blue', label='free_clean', linestyle='-', marker='o')
plt.plot(free_x, free_adv, color='blue', label='free_adv', linestyle=':', marker='o')

plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
```

## 训练耗时

```python
plt.figure(figsize=(8, 5))

plt.plot(fgsm_x, fgsm_time, color='red', label='fgsm', linestyle='-', marker='s')

plt.plot(pgd_x, pgd_time, color='green', label='pgd', linestyle='-', marker='^')

plt.plot(free_x, free_time, color='blue', label='free', linestyle='-', marker='o')

plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Training time (s)')
```
